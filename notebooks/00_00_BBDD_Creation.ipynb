```python
import os
import sys
import numpy as np
import pandas as pd

from os.path import join

pd.set_option("display.max_columns", None)
pd.options.display.float_format = '{:,.2f}'.format
```

## Cargamos BBDD de Falsos Reemplazos - SAS

```python
path = r"C:\Users\e107829\Allianz\Price to Market - Turismos - Los documentos\17. Reemplazos y Falsos Reemplazos\Decision_Tree"
```

```python
BBDD_FR = pd.read_csv(os.path.join(path, "ANALISIS.csv"), sep=';', encoding='latin-1')

BBDD_FR.head(5)
```

```python
len(BBDD_FR)
```

```python
BBDD_FR['FRENOV'] = pd.to_datetime(BBDD_FR['FRENOV'])
```

#### Obtenemos unicamente las polizas de 2023

```python
# Filtrar por 2023 FRENOV

df_23 = BBDD_FR[BBDD_FR['FRENOV'].dt.year == 2023]
len(df_23)
```

## Cargamos BBDD de Turismos de FR

```python
# Unificar creando PFR

# DF2 = pd.read_csv(os.path.join(path, "Turismos_v3.csv"), sep=';', encoding='latin-1')
DF2 = pd.read_csv(os.path.join(path, "FRPs Diciembre Franq arreglado y respuesta tirea.csv"), sep=';', encoding='latin-1')
DF2.head(5)
```

#### Creamos la variable Falsos Reemplazos

```python
FR = DF2[DF2.FRFINAL == 1]
len(FR)
```

```python
PFR = df_23['POLIZA'].isin(FR['POLIZA'])

df_23.loc[PFR, 'FRFINAL'] = 1
df_23.loc[~PFR, 'FRFINAL'] = 0

df_23.FRFINAL.value_counts()
```

## Cargamos BBDD de Garantias e incluimos ICORR

```python
# Unificiar con IC

DF3 = pd.read_csv(os.path.join(path, "GARANTIAS_v2.csv"), sep=';', encoding='latin-1')
DF3.head(5)
```

```python
len(DF3)
```

```python
for gar_num in sorted(DF3.GARANTIA.unique()):
    DF3.loc[DF3["GARANTIA"] == gar_num, f"GARANTIA_{gar_num}"] = DF3.loc[DF3["GARANTIA"] == gar_num, "ICORR"]
```

```python
DF3.head(2)
```

```python
df_grouped = DF3[["POLIZA", "RAMO", "GARANTIA_1", "GARANTIA_2", "GARANTIA_3", "GARANTIA_4",
                  "GARANTIA_5", "GARANTIA_6", "GARANTIA_7", "GARANTIA_8"]].groupby(by=["POLIZA", "RAMO"]).agg('max').reset_index()
df_grouped
```

```python
DF_VF = pd.merge(df_23, df_grouped, on=['POLIZA', 'RAMO'], how='left')
DF_VF.head(2)
```

```python
print(len(df_23))
print(len(df_grouped))
print(len(DF_VF))
```

```python
Garantias = ['GARANTIA_1', 'GARANTIA_2', 'GARANTIA_3', 'GARANTIA_4', 'GARANTIA_5', 'GARANTIA_6', 'GARANTIA_7', 'GARANTIA_8']
#DF_VF['Gar_missing'] = DF_VF[DF_VF.columns[86:]].isnull().sum(axis=1)
DF_VF['Gar_missing'] = DF_VF[Garantias].isnull().sum(axis=1)
DF_VF.head(2)
```

```python
DF_VF.Gar_missing.value_counts()
```

```python
len(DF_VF[DF_VF.CODPOST == '*****'])
```

```python
DF_VF.loc[DF_VF.CODPOST == '*****', 'CODPOST'] = np.nan
```

```python
DF_VF.loc[DF_VF.Edad_conductor == '**', 'Edad_conductor'] = np.nan
```

### Tratamos ciertas columnas para evitar errores al guardar el PQ.

```python
DF_VF['CODPOST'] = DF_VF['CODPOST'].astype('float32')
DF_VF['USO'] = DF_VF['USO'].astype(str).astype(object)
DF_VF['sdad_total3p'] = DF_VF['sdad_total3p'].astype(str).astype(object)
DF_VF['sdad_total_tom'] = DF_VF['sdad_total_tom'].astype(str).astype(object)
DF_VF['Edad_conductor'] = DF_VF['Edad_conductor'].astype('float32')
DF_VF['Provincia'] = DF_VF['Provincia'].astype(str).astype(object)
```

```python
DF_VF.to_parquet(os.path.join(path, "BBDD_FR.parquet"))
```
